```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct  1 22:20:51 2019

@author: ankur sinha
"""

import pandas as pd
from sklearn.model_selection import train_test_split
from numpy import array
from numpy import argmax
from sklearn.linear_model import LinearRegression

#loading all csv to dataframes
df_bluru = pd.read_csv("team_stats_bengaluru.csv")
df_bbulls = pd.read_csv("team_stats_bbulls.csv")
df_delhi = pd.read_csv("team_stats_delhi.csv")
df_guj = pd.read_csv("team_stats_gujrat.csv")
df_haryana = pd.read_csv("team_stats_haryana.csv")
df_jaipur =  pd.read_csv("team_stats_jaipur.csv")
df_mumba = pd.read_csv("team_stats_delhi.csv")
df_patna = pd.read_csv("team_stats_patna.csv")
df_pune = pd.read_csv("team_stats_pune.csv")
df_tamil = pd.read_csv("team_stats_tamil.csv")
df_telugu = pd.read_csv("team_stats_telugu.csv")
df_up = pd.read_csv("team_stats_up.csv")

headers = df_bluru.iloc[:,1]

print(df_bluru.dtypes)
df_bluru["aggr"] = (df_bluru["Season 1"]+df_bluru["Season 2"]+df_bluru["Season 3"]+df_bluru["Season 4"]+df_bluru["Season 5"]+df_bluru["Season 6"])/6
print(df_bluru["aggr"])
df_bbulls["aggr"] = (df_bbulls["Season 1"]+df_bbulls["Season 2"]+df_bbulls["Season 3"]+df_bbulls["Season 4"]+df_bbulls["Season 5"]+df_bbulls["Season 6"])/6
print(df_bbulls["aggr"])
df_delhi["aggr"] = (df_delhi["Season 1"]+df_delhi["Season 2"]+df_delhi["Season 3"]+df_delhi["Season 4"]+df_delhi["Season 5"]+df_delhi["Season 6"])/6
df_jaipur["aggr"] = (df_jaipur["Season 1"]+df_jaipur["Season 2"]+df_jaipur["Season 3"]+df_jaipur["Season 4"]+df_jaipur["Season 5"]+df_jaipur["Season 6"])/6
df_mumba["aggr"] = (df_mumba["Season 1"]+df_mumba["Season 2"]+df_mumba["Season 3"]+df_mumba["Season 4"]+df_mumba["Season 5"]+df_mumba["Season 6"])/6
df_patna["aggr"] = (df_patna["Season 1"]+df_patna["Season 2"]+df_patna["Season 3"]+df_patna["Season 4"]+df_patna["Season 5"]+df_patna["Season 6"])/6
df_pune["aggr"] = (df_pune["Season 1"]+df_pune["Season 2"]+df_pune["Season 3"]+df_pune["Season 4"]+df_pune["Season 5"]+df_pune["Season 6"])/6
df_telugu["aggr"] = (df_telugu["Season 1"]+df_telugu["Season 2"]+df_telugu["Season 3"]+df_telugu["Season 4"]+df_telugu["Season 5"]+df_telugu["Season 6"])/6
df_guj["aggr"] = (df_guj["Season 5"]+df_guj["Season 6"])/2
df_haryana["aggr"] = (df_haryana["Season 5"]+df_haryana["Season 6"])/2
df_tamil["aggr"] = (df_tamil["Season 5"]+df_tamil["Season 6"])/2
df_up["aggr"] = (df_up["Season 5"]+df_up["Season 6"])/2
     
df = pd.DataFrame(columns=headers)
df.loc[0] = list(df_bluru["aggr"])
df.loc[1] = list(df_bbulls["aggr"])
df.loc[2] = list(df_delhi["aggr"])
df.loc[3] = list(df_jaipur["aggr"])
df.loc[4] = list(df_mumba["aggr"])
df.loc[5] = list(df_patna["aggr"])
df.loc[6] = list(df_pune["aggr"])
df.loc[7] = list(df_telugu["aggr"])
df.loc[8] = list(df_guj["aggr"])
df.loc[9] = list(df_haryana["aggr"])
df.loc[10] = list(df_tamil["aggr"])
df.loc[11] = list(df_up["aggr"])
print(df)
df.to_csv("teams_overall_data.csv")
                        
```
